```python
items = [
    "비누", "치약", "샴푸", "린스", "바디워시", "폼클렌징", "칫솔", "수건",
    "휴지", "물티슈", "세탁세제", "섬유유연제", "주방세제", "수세미", "고무장갑",
    "쌀", "라면", "햇반", "생수", "우유", "계란", "두부", "콩나물", "시금치",
    "양파", "감자", "고구마", "사과", "바나나", "오렌지", "귤", "토마토",
    "김치", "된장", "고추장", "간장", "식용유", "참기름", "소금", "설탕",
    "커피", "차", "과자", "빵", "젤리", "초콜릿", "음료수", "맥주", "소주",
    "고기(돼지고기)", "고기(소고기)", "닭고기", "생선", "오징어", "새우", "게",
    "쌀국수", "파스타", "잼", "버터", "치즈", "요거트", "아이스크림", "통조림",
    "냉동만두", "어묵", "햄", "소시지", "김", "미역", "다시마", "멸치",
    "밀가루", "부침가루", "튀김가루", "빵가루", "식초", "소스", "향신료",
    "양초", "성냥", "건전지", "전구", "쓰레기봉투", "지퍼백", "호일", "랩"
]

#가격 랜덤설정
import random
prices = {item: random.randint(1000, 10000) for item in items} #random.randint 1000~10000 랜덤 정수 생성

#재고 초기 10개씩
stock = {item: 10 for item in items}

#장바구니
cart = {} #담은 상품 저장공간 처음에는 비어있음

#매출
total_sales = 0 #총 매출 금액 저장 처음에는 0

#메뉴
while True: #반복문
    print("\n---- 마트 프로그램 ----")
    print("1. 상품 보기")
    print("2. 장바구니 담기")
    print("3. 장바구니 확인 및 결제")
    print("4. 재고 확인")
    print("5. 매출 확인")
    print("0. 종료") #메뉴 출력

    choice = input("메뉴 선택: ") #input 사용자로부터 값을 입력 받음

    #1. 상품 보기
    if choice == "1": #메뉴 1 선택 시 실행
        for i, item in enumerate(items): #enumerate 리스트를 (번호 i, 값 item) 형태로 만듦
            print(f"{i}. {item} - {prices[item]}원 (재고: {stock[item]})") #상품번호, 이름, 가격, 재고 출력

    #2. 장바구니 담기
    elif choice =="2":
        index = int(input("상품 번호 입력: ")) #상품번호 입력받고 숫자로 변환
        if index < 0 or index >= len(items): #len 리스트 길이 반환, 잘못된 번호입력 체크
            print("잘못된 번호입니다.")
            continue #잘못된 입력이면 다시 메뉴로

        item = items[index] #선택한 상품번호 = 상품 이름으로 변환
        qty = int(input("수량 입력: ")) #int 문자열 -> 정수 변환, 구매할 수량 입력

        if stock[item] < qty: #재고보다 많이 주문했는지 확인
            print("재고 부족")
        else: #재고 충분 시 실행
            cart[item] = cart.get(item, 0) + qty #cart.get 카트에서 값 가져오기, 장바구니 상품 추가(기존에 있으면 더하기, 없으면 0에서 시작)
            stock[item] -= qty #재고 감소
            print(f"{item} {qty}개 장바구니에 담김")

    #3. 결제
    elif choice == "3":
        if not cart: #장바구니가 비었는지 확인
            print("장바구니가 비어있습니다.")
            continue #비어있으면 다시 메뉴

        total = 0 #총 금액 초기화
        print("\n--- 장바구니 ---")
        for item, qty in cart.items(): #장바구니에 있는 상품 반복
            cost = prices[item] * qty #상품 가격 * 수량
            print(f"{item} x {qty} = {cost}원") #계산 결과 출력
            total += cost #총 금액에 더하기

        print(f"총 금액: {total}원") #총 금액 출력

        pay = input("결제하시겠습니까? (y/n): ")
        if pay == "y": #결제 진행
            total_sales += total #매출 증가
            cart.clear() #cart.clear 카트 비우기
            print("결제 완료")
        else:
            print("결제 취소")

    #4. 재고 확인
    elif choice == "4":
        print("\n--- 재고 현황 ---")
        for item in items: #모든 상품 반복
            print(f"{item}: {stock[item]}개") #재고 출력

    #5. 매출 확인
    elif choice == "5":
        print(f"\n총 매출: {total_sales}원") #지금까지의 매출 출력

    #0. 종료
    elif choice == "0":
        print("프로그램 종료")
        break #break 반복문 종료

    else:
        print("잘못된 입력입니다.") #0~5 외 입력 처리
```
