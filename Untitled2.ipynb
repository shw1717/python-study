```python
num = int(input("숫자 입력: ")) # 정수 입력 받기

list_1 = [] # 약수 저장할 리스트
for i in range(1, num+1): # 1부터 num까지 반복하면서 약수 찾기
    if num % i == 0:
        list_1.append(i)

print(f"{num} : {','.join(map(str,list_1))}") #결과 출력
```

```python
num = int(input("정수 입력: "))

print("{}의 약수 : ".format(num), end='')

for i in range(1, num + 1):
    if num % i != 0:
        continue

    if i == num:
        print(i)
        break

    print(f"{i}", end=', ')
```

```python
print("{}의 약수 : 1, ".format(num), end='')

i = 2
while True:
    if num % i == 0:
        print("{},".format(i), end=' ')

    i = i+1

    if i == num:
        break

print(num)
```
