# 1. 파이썬이란?
파이썬은 사람이 이해하기 쉬운 문법을 가진 인터프리터 언어이다.  
중괄호 대신 들여쓰기(indent)로 코드 블록을 구분하며, 세미콜론(`;`)이 필요하지 않다.

```python
print("Hello, Python!")
print("파이썬 첫 실행입니다.")
```

# 2. 주석(Comment)
주석은 프로그램 실행에는 영향을 주지 않으며, 코드를 설명하기 위해 사용된다.  
주석은 `#`으로 작성한다.

```python
# 이것은 한 줄 주석입니다.
print("주석은 코드 설명에 사용됩니다.") # 코드 끝에 주석 추가 가능
```

# 3. 들여쓰기(Indentation)
파이썬은 들여쓰기로 코드 블록을 구분한다.  
일반적으로 공백 4칸을 사용하며, 들여쓰기 오류 시 `IndentationError`가 발생한다.

```python
# @title
score = 85

if score >= 80:
  print("좋은 점수입니다!")
  print("축하합니다.")
else:
  print("조금 더 노력해요!")
  print("아직부족합니다.")
```

# 4. 변수(Variable)
변수는 데이터를 저장하기 위한 이름표이다.  
`변수명 = 값` 형태로 사용하며, 대소문자를 구분한다.

```python
name = "홍길동"
age = 25
height = 175.5

print(name)
print(age)
print(height)
```

# 6. 문자열(String)


*   작은따옴표('), 큰따옴표(") 모두 사용 가능
*   문자열 연결: +, 반복: *

# 7. 출력(print)
`print()` 함수는 화면에 결과를 출력한다.  
여러 변수를 한 번에 출력하거나, `f-string` 형식을 이용하면 가독성이 높다.

```python
name = "홍길동"
age = 25
print("이름:", name, ", 나이:", age)
print(f"이름은 {name}, 나이는 {age}살입니다.")
```

# 8. 입력(input)
`input()` 함수는 사용자에게 메시지를 보여주고 입력을 받을 수 있게 한다.  
항상 문자열(str) 형태로 반환되며, 숫자형으로 사용하려면 `int()` 또는 `float()`로 형 변환이 필요하다.

```python
name = input("이름을 입력하세요:")
print("안녕하세요,", name, "님!")
```

# 입력 데이터 변환 예시
`input()`은 기본적으로 문자열을 반환하기 때문에, 숫자 계산을 위해서는 형변환이 필요하다.

```python
age = int(input("나이를 입력하세요: "))
print(f"10년 후 나이는 {age + 10}살입니다.")
```

# 9. 기본 연산자
파이썬의 기본 산술 연산자는 다음과 같습니다.
| 연산자 | 설명 | 예시 | 결과 |
|---|---|---|---|
| + | 덧셈 | 5 + 3 | 8 |
| - | 뺄셈 | 5 - 3 | 2 |
| * | 곱셈 | 5 * 3 | 15 |
| / | 나눗셈(실수형) | 5 / 2 | 2.5 |
| // | 몫(정수형) | 5 // 2 | 2 |
| % | 나머지 | 5 % 2 | 1 |
| ** | 거듭제곱 | 2 ** 3 | 8 |

※ `+`, `*`는 문자열에도 사용할 수 있습니다.

```python
a, b = 5, 2
print("덧셈:", a + b)
print("나눗셈:", a / b)
print("몫:", a // b)
print("나머지:", a % b)
print("거듭제곱:", a ** b)

# 문자열 연산 예시
print("Hello" + "Python")
print("Hi!" * 3)
```

## 10. 비교 및 논리 연산자

### 🔸 비교 연산자
두 값을 비교하여 **참(True)** 또는 **거짓(False)** 을 반환합니다.

| 연산자 | 의미 | 예시 | 결과 |
|---|---|---|---|
| == | 같다 | 5 == 5 | True |
| != | 다르다 | 5 != 3 | True |
| > | 크다 | 5 > 3 | True |
| < | 작다 | 5 < 3 | False |
| >= | 크거나 같다 | 5 >= 5 | True |
| <= | 작거나 같다 | 3 <= 5 | True |



### 🔸 논리 연산자
여러 조건을 결합하거나 뒤집을 때 사용합니다.

| 연산자 | 의미 | 예시 | 결과 |
|---|---|---|---|
| and | 두 조건이 모두 참일 때 | (3 > 1) and (2 < 5) | True |
| or | 하나라도 참이면 | (3 < 1) or (2 < 5) | True |
| not | 참/거짓을 반대로 | not(3 > 1) | False |

```python
x, y = 10, 5
print(x > y)
print(x == y)
print(x != y)

print((x > y) and (x < 20))
print((x < y) or (x == 10))
print(not (x > y))
```
