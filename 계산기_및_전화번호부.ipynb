1. 계산기 - 리스트 입력받음(숫자, 문자 예외처리) -> +는 다 더하기, -는 다 빼기, *는 다 곱하기, /는 다 나누기 -> 평균값, 최대값, 최소값 구하기 class를 이용하기

2. 전화번호부 - 개인: 이름, 이메일주소, 전화번호, 직장명... -> 사용자가 개인정보 알려주면 저장, 필요하면 불러오기, 검색기능 class를 활용하여 만들기

```python
class Calculator:
    def __init__(self, data): #초기화
        self.data = data #사용자가 입력한 리스트
        self.numbers = [] #숫자
        self.operators = [] #연산자
        self.result = None #계산결과

    def parse_input(self): #입력값 분리
        for item in self.data: #반복문
            if isinstance(item, (int, float)): #숫자인 경우 numbers 리스트에 저장
                self.numbers.append(item)
            elif item in ['+', '-', '*', '/']: #연산자인 경우 operators 리스트에 저장
                self.operators.append(item)
            else:
                raise ValueError(f"잘못 입력: {item}") #둘다 아닐시 에러

        if len(self.numbers) - 1 != len(self.operators):
            raise ValueError("연산자 개수가 숫자보다 하나 적어야 함.")

    def calculate(self): #연산 수행
        self.result = self.numbers[0] #첫번째 숫자 기준 시작

        for i in range(len(self.operators)): #연산자 개수만큼 반복
            op = self.operators[i]
            num = self.numbers[i+1] #연산자 + 다음 숫자 가져오기

            if op == '+':
                self.result += num
            elif op == '-':
                self.result -= num
            elif op == '*':
                self.result *= num
            elif op == '/': #연산자에 따라 계산
                if num == 0:
                    raise ZeroDivisionError("0으로 나눌 수 없음.")
                self.result /= num

        return self.result #결과 반환

    def get_stats(self): #통계 함수
        return {
            "평균": sum(self.numbers) / len(self.numbers),
            "최대값": max(self.numbers),
            "최소값": min(self.numbers)
        } #sum=총합, len=개수, max=최대, min=최소
try:
    user = input("수식 입력:")

    tokens = user.split() #문자열 쪼개기
    converted = [] #빈 리스트 생성

    for t in tokens:
        if t.isdigit(): #문자열이 숫자인지 검사
            converted.append(int(t)) #문자열 -> 숫자 변환
        else:
            converted.append(t) #연산자는 유지

    calc = Calculator(converted)

    calc.parse_input() #입력 검사
    result = calc.calculate() #계산 수행
    stats = calc.get_stats() #평균,최대,최소 계산

    print("계산 결과:", result)
    print("평균:", stats["평균"])
    print("최대값:", stats["최대값"])
    print("최소값:", stats["최소값"]) #결과 출력

except Exception as e:
    print("에러 발생:", e)
```

```python
class Contact: #한 명의 정보를 저장하는 클래스
    def __init__(self, name, email, phone, company):
        self.name = name #이름
        self.email = email #이메일
        self.phone = phone #전화번호
        self.company = company #직장

    def __str__(self):
        return f"이름: {self.name}, 이메일: {self.email}, 전화번호: {self.phone}, 직장: {self.company}"


class PhoneBook:
    def __init__(self):
        self.contacts = []  # 연락처 저장 리스트

    def add_contact(self): #사용자 입력 받음
        name = input("이름: ")
        email = input("이메일: ")
        phone = input("전화번호: ")
        company = input("직장명: ")

        contact = Contact(name, email, phone, company)
        self.contacts.append(contact) #리스트에 저장

        print("연락처가 저장되었습니다.\n")

    def show_all(self): #저장된 연락처 출력
        if not self.contacts:
            print("저장된 연락처가 없습니다.\n")
            return

        print("전체 연락처 목록")
        for c in self.contacts:
            print(c)
        print()

    def search(self): #이름 검색
        keyword = input("검색할 이름: ")

        found = False
        for c in self.contacts:
            if keyword == c.name:
                print(c)
                found = True

        if not found:
            print("검색 결과 없음\n")
        else:
            print()

    def run(self):
        while True:
            print("==== 전화번호부 ====")
            print("1. 연락처 추가")
            print("2. 전체 조회")
            print("3. 검색")
            print("4. 종료")

            choice = input("선택: ")

            if choice == '1':
                self.add_contact()
            elif choice == '2':
                self.show_all()
            elif choice == '3':
                self.search()
            elif choice == '4':
                print("프로그램 종료")
                break
            else:
                print("잘못된 입력입니다.\n")

phonebook = PhoneBook()
phonebook.run()
```
