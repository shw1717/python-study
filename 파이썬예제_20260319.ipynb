```python
print("{}의 약수 : 1, ".format(num), end='')

i = 2
while True:
    if num % i == 0:
        print("{}, ".format(i), end='')

    i = i + 1

    if i == num:
        break

print(num)
```

```python
print("{}의 약수 : 1, ".format(num), end='')

i = 2
while i < num: # i != num
    if num % i != 0:
        i = i + 1
        continue

    print("{}, ".format(i), end='')
    i = i + 1

print(num)
```

```python
def 약수구하기(숫자):
    print("{}의 약수 : 1, ".format(숫자), end='')
    #print(num, "의 약수", sep='')

    for i in range(2, 숫자):
        if 숫자 % i == 0:
            print("{}, ".format(i), end='')

    print(숫자)
```

```python
def 약수구하기_방법2(숫자, 화면출력=False):
    약수들 = [1]
    for i in range(2, 숫자):
        if 숫자 % i == 0:
            #약수들 = 약수들 + [i]
            약수들.append(i)
    약수들.append(숫자)

    if 화면출력:
        print("{}의 약수 : ".format(숫자), end='')
        for 약수 in 약수들[0:len(약수들)-1]:
            print("{}, ".format(약수), end='')
        print(약수들[-1])

    return 약수들
```

```python
while True:
    num = int(input("약수를 구하기 위한 정수를 입력해 주세요: "))
    #약수구하기(num)
    약수들 = 약수구하기_방법2(num, True)

    rep = input("더이상 약수를 구하지 않을거면 'y'를 입력해주세요: ")
    if rep == 'y':
        break
```
