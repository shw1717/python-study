문제) 컴퓨터와 가위바위보 게임을 코딩해보세요. 5전 3선승제로 먼저 이기는 사람을 알려주고, 축하의 인사도 출력해 보세요

```python
import random

user = 0
com = 0

while user < 3 and com < 3:
    u = int(input("가위(1) 바위(2) 보(3): "))
    c = random.randint(1, 3)

    print("컴퓨터:", c)

    if u == c:
        print("무승부")
    elif (u == 1 and c == 3) or (u == 2 and c == 1) or (u == 3 and c == 2):
        print("승리!")
        user += 1
    else:
        print("패배!")
        com += 1

    print("점수:", user, com)

if user == 3:
    print("최종 승리! 축하합니다!")
else:
    print("컴퓨터 승리!")
```
